```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")

abb = pd.read_csv('AB_US_2020.csv')
```

```python
abb = abb.dropna(axis = 0)
```

```python
abb.columns
```

```python
#plot before removing outliers

# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Airbnb Prices vs Minimum Nights")

sns.scatterplot(x=abb['minimum_nights'], y=abb['price'])
```

```python
#removing unlikely outliers for price and minimum nights

abb = abb[(abb['price'] > 75) & (abb['price'] < 1000)]
abb = abb[(abb['minimum_nights'] < 30)]
```

```python
#plot after removing outliers

# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Airbnb Prices vs Minimum Nights")

sns.scatterplot(x=abb['minimum_nights'], y=abb['price'])
```

```python
plt.figure(figsize = (10, 6))
plt.title('"Airbnb Prices vs Number of Reviews"')
sns.scatterplot(x = abb['number_of_reviews'], y = abb['price'])
```

```python
min_num_reviews = abb.number_of_reviews.min()
max_num_reviews = abb.number_of_reviews.max()
print(min_num_reviews, max_num_reviews)
#no apparent outliers(min = 1, max = 827)
```
